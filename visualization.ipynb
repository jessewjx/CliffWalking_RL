```python
import MDP_solver
import numpy as np
import matplotlib.pyplot as plt
```

## Fixed Epsilon

```python
# initial condition
varying_epsion = False
epsilon = 0.1
num_episodes = 500
num_runs = 10
gamma = 1
alpha = 0.5
xdim = 12
ydim = 4
gw = MDP_solver.Solver(epsilon, num_episodes, alpha, gamma, xdim, ydim, varying_epsion)
```

```python
# Sarsa learning
Sreward_arr=np.zeros(num_episodes)
for r in range(0,num_runs):
    Sreward_trj = gw.Sarsa()
    Sreward_arr = Sreward_arr + np.array(Sreward_trj)
Sreward_arr = Sreward_arr/float(num_runs)

print("Visualization of Sarsa policy")
gw.Visualize_learned_policy()
```

```python
# plot average cummulative reward as a function of number of episodes
x = [i for i in range(0, num_episodes)]
plt.plot(x,Qreward_arr,label = "Q learning")
plt.plot(x, Sreward_arr,label = "Sarsa")
plt.xlabel('episode count')
plt.ylabel('cummulative reward')
plt.ylim([-200, 0])
plt.legend()
plt.show()
```

3. Both Q learning and Sarsa is sampling from the full DP tree to estimate the Q value of a state, action pair. However Q learning is off policy while Sarsa is on policy, which is learning the global optimal policy/performing policy improvement by taking the action that maximizes the accumulative reward (immediate reward plus value function of successor state) while Sarsa is on policy, which is learning the action function of epsilon greedy policy by taking local greedy action that maximizes the accumulative reward. 

## Varying Epsilon

```python
# initial condition
varying_epsilon = True
epsilon = 0.1
num_episodes = 500
num_runs = 10
gamma = 1
alpha = 0.5
xdim = 12
ydim = 4
gw = MDP_solver.Solver(epsilon, num_episodes, alpha, gamma, xdim, ydim,varying_epsilon)
```

```python
# Q learning
Qreward_arr=np.zeros(num_episodes)
# average the episodes rewards over runs
for r in range(0,num_runs):
    Qreward_trj = gw.Qlearning()
    Qreward_arr = Qreward_arr + np.array(Qreward_trj)
Qreward_arr = Qreward_arr/float(num_runs)

print("Visualization of Qlearning policy")
gw.Visualize_learned_policy()
```

```python
# Sarsa learning
Sreward_arr=np.zeros(num_episodes)
for r in range(0,num_runs):
    Sreward_trj = gw.Sarsa()
    Sreward_arr = Sreward_arr + np.array(Sreward_trj)
Sreward_arr = Sreward_arr/float(num_runs)

print("Visualization of Sarsa policy")
gw.Visualize_learned_policy()
```

```python
# plot average cummulative reward as a function of number of episodes
x = [i for i in range(0, num_episodes)]
plt.plot(x,Qreward_arr,label = "Q learning")
plt.plot(x, Sreward_arr,label = "Sarsa")
plt.xlabel('episode count')
plt.ylabel('cummulative reward')
plt.ylim([-200, 0])
plt.legend()
plt.show()
```
