```python
import MDP_solver
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Q learning
def Qlearning():
    print("start Q learning...")
    Qreward_arr=np.zeros(num_episodes)
    # average the episodes rewards over runs
    for r in range(0,num_runs):
        Qreward_trj = gw.Qlearning()
        Qreward_arr = Qreward_arr + np.array(Qreward_trj)
    Qreward_arr = Qreward_arr/float(num_runs)

    print("Visualization of Qlearning policy")
    gw.Visualize_learned_policy()
    
    return Qreward_arr

# Sarsa learning
def Sarsa_learning():
    print("\n\nstart Sarsa learning...")
    Sreward_arr=np.zeros(num_episodes)
    for r in range(0,num_runs):
        Sreward_trj = gw.Sarsa()
        Sreward_arr = Sreward_arr + np.array(Sreward_trj)
    Sreward_arr = Sreward_arr/float(num_runs)

    print("Visualization of Sarsa policy")
    gw.Visualize_learned_policy()
    
    return Sreward_arr

# plot average cummulative reward as a function of number of episodes
def plot_cummulative_reward(Qreward_arr, Sreward_arr):
    x = [i for i in range(0, num_episodes)]
    plt.plot(x,Qreward_arr,label = "Q learning")
    plt.plot(x, Sreward_arr,label = "Sarsa")
    plt.xlabel('episode count')
    plt.ylabel('cummulative reward')
    plt.ylim([-500, 0])
    plt.legend()
    plt.show()
```

## Fixed Epsilon

```python
# initial condition
varying_epsion = False
initial_Q = 0
epsilon = 0.1
num_episodes = 500
num_runs = 10
gamma = 1
alpha = 0.5
xdim = 12
ydim = 4
gw = MDP_solver.Solver(epsilon, num_episodes, alpha, gamma, xdim, ydim, varying_epsion,initial_Q)
```

```python
Qreward=Qlearning()
Sreward=Sarsa_learning()
plot_cummulative_reward(Qreward,Sreward)
```

3. Both Q learning and Sarsa is sampling from the full DP tree to estimate the Q value of a state, action pair. However Q learning is off policy, which is learning the global optimal policy/performing policy improvement by taking the action that maximizes the accumulative reward (immediate reward plus value function of successor state) while Sarsa is on policy, which is learning the action function of epsilon greedy policy by taking local greedy action that maximizes the accumulative reward. 

## Bonus Q function initialization

```python
# initial condition
varying_epsilon = False
initial_Q = -1000
epsilon = 0.1
num_episodes = 500
num_runs = 10
gamma = 1
alpha = 0.5
xdim = 12
ydim = 4
gw = MDP_solver.Solver(epsilon, num_episodes, alpha, gamma, xdim, ydim,varying_epsilon,initial_Q)
```

```python
Qreward=Qlearning()
Sreward=Sarsa_learning()
plot_cummulative_reward(Qreward,Sreward)
```

When Q is initialized to -1000, both Q learning and sarsa converges faster probably because most of the actions have negative reward, while starting with 0, due to in the DP tree, the recurrence is a max operation over all the successor states so the action function of most of the action function will not be updated while starting with -1000, the value of a state can be learned and updated directly.

## Varying Epsilon

```python
# initial condition
varying_epsilon = True
initial_Q = 0
epsilon = 0.1
num_episodes = 500
num_runs = 10
gamma = 1
alpha = 0.5
xdim = 12
ydim = 4
gw = MDP_solver.Solver(epsilon, num_episodes, alpha, gamma, xdim, ydim,varying_epsilon,initial_Q)
```

```python
Qreward=Qlearning()
Sreward=Sarsa_learning()
plot_cummulative_reward(Qreward,Sreward)
```

When the exploration parameter is decreasing, the learning process is gradually of pure exploitation, both Q learning and Sarsa is ultimating solving the estimated MDP thus the cummulative reward of both is the same.
